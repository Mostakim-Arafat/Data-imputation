```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv("DD.csv")
df.head()
```

```python
df.info()
```

```python
df2=df.copy()
df2=df2.drop(['Pregnant','Glucose','Skin_Fold','Serum_Insulin','Diabetes_Pedigree','Age','Class'],axis=1)
df2
```

```python
model=LinearRegression()
#first do train non null dataset
df_train=df2.dropna(subset=['BMI','Diastolic_BP'])
#now train data
model.fit(df_train[['BMI']],df_train['Diastolic_BP'])
#NOW WANT TO PREDICT
#SO FIRST CHECK WHERE DIASTOLIC IS NULL
df_p=df2.loc[df2['Diastolic_BP'].isnull(),['BMI']]
#ENSURE BMI NOT NULL
df_p_1=df_p.dropna(subset=['BMI'])
df_p_1.head()
```

```python
# PREDICT DIASTOLIC BP
df_predict=model.predict(df_p_1[['BMI']])
update_index=df_p_1.index
#here we set updated Diastolic_BP in original dataset df2
df2.loc[update_index,'Diastolic_BP']=df_predict
df2
```
